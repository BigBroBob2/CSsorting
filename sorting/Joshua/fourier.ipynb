```python
import os
import numpy as np
import matplotlib.pyplot as plt

os.chdir('/Users/xiao-zy19/Desktop/good_cs')
data = np.fromfile('temp_wh.dat', dtype=np.int16).reshape(-1, 128) #reshape((-1, 128))
print(data.shape)
print(data[:10, 23])
```

```python
channel23 = data[:, 23]

complex_spike_time = [72.4835, 72.5615, 93.076, 113.5851, 134.1635, 134.167, 134.1768, 134.2044, 134.2447, 134.7556, 134.7862, 134.9443, 134.9715, 135.0115, 135.1218, 135.1738, 135.2197] #  93.076,
complex_spike_index = [int(x*30000) for x in complex_spike_time]
```

```python
for i in complex_spike_index:
    # FFT
    segment = channel23[i-100:i+200]  # extract data points
    fft_segment = np.fft.fft(segment)  # FFT
    freqs = np.fft.fftfreq(len(segment), 1/30000) # fs=30000Hz

    # plot frequency spectrum
    plt.figure(figsize=(12, 6))
    plt.plot(freqs, np.abs(fft_segment))  # np.abs
    plt.title(f"Frequency Spectrum around Spike index {i}")
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.xlim(0, 15000)
    plt.grid(linestyle='--', alpha=0.5)
    plt.show()
```
